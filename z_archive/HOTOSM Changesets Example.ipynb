```python
import pandas as pd
%matplotlib inline
```

```python
df = pd.read_csv('./data/last_100k_hotosm_changesets.csv')
```

```python
df['date'] = df.created_at.apply(lambda x: pd.Timestamp(x).date())
```

```python
df.head(2)
```

```python
df.groupby('date').aggregate({'id':'count'}).plot(figsize=(15,8))
```

```python
df.groupby('user').aggregate('count').sort_values(by='id',ascending=False)
```

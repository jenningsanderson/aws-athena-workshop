```python
df = pd.read_csv('/Users/jenningsanderson/Downloads/af8247c3-2217-4ea2-8430-e593e879c46e.csv')
```

```python
df.groupby('user').aggregate('nunique').sort_values(by='id',ascending=False).head(25)
```

```python
df['date'] = df.timestamp.apply(lambda x: pd.Timestamp(x).date())
```

```python
df.groupby('date').aggregate('nunique')['user'].plot(figsize=(15,8))
```

```python
df.groupby('date').aggregate('nunique')[['user']].plot(figsize=(15,8))
```

```python
df.groupby('date').aggregate('count')['id'].plot(kind='bar',stacked=True,figsize=(15,8))
```

```python
df['is_hot'] = df['tags.1'].apply(lambda x: 'hotosm' in x)
df['is_fix'] = df['tags.1'].apply(lambda x: 'fix' in x)
df['is_validation'] = df['tags.1'].apply(lambda x: 'valid' in x)
df['is_2770'] = df['tags.1'].apply(lambda x: '2770' in x)
```

```python
df.groupby('date').aggregate('sum')[['is_hot','is_validation','is_fix']].plot(kind='bar',stacked=True,figsize=(15,8))
                                      
```

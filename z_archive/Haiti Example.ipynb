```python
def string_to_dict(string):
    if type(string)==str:
        arr = string[1:-1].split("=")
        d = dict()
        for k, v in zip(arr[:-1], arr[1:]):
            d[k] = v
        return d
    elif type(string)==dict:
        return string
```

```python
# Convert the tags from strings to dictionaries
df["tags"] = df.tags.apply(string_to_dict)
df["c_tags"] = df.tags.apply(string_to_dict)
```

```python
# Convert the timestamps to timestamps
df['updated'] = df.updated.apply(pd.Timestamp)
df['valid_until'] = df.valid_until.apply(pd.Timestamp)
```

```python
df['lifespan'] = df.apply(lambda row: row.valid_until - row.updated, axis=1)
```
